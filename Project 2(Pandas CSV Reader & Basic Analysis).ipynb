```python
import pandas as pd

# Step 1: Read CSV File
df = pd.read_csv("covid19.csv")
df

```

```python
# Step 2: Display First & Last Rows
print("First 5 Rows:\n")
print(df.head())

print("\nLast 5 Rows:\n")
print(df.tail())

```

```python
# Step 3: Check Data Types
print("\nData Types:\n")
print(df.dtypes)
```

```python
# Step 4: Summary Statistics
print("\nSummary Statistics:\n")

print("Mean:\n", df.mean(numeric_only=True))
print("\nMedian:\n", df.median(numeric_only=True))
print("\nMinimum:\n", df.min())
print("\nMaximum:\n", df.max())
print("\nCount:\n", df.count())
```

```python
# Step 5: Convert Date Column
df['ObservationDate'] = pd.to_datetime(df['ObservationDate'])
print("\nDate Column Converted Successfully!")
```

```python
# Step 6: Filter Rows (Confirmed Cases > 1000)
filtered_data = df[df['Confirmed'] > 1000]

print("\nFiltered Data (Confirmed > 1000):\n")
print(filtered_data)
```

```python
# Step 6: Filter Rows (Confirmed Cases > 1000)
filtered_data = df[df['Confirmed'] > 1000]

print("\nFiltered Data (Confirmed > 1000):\n")
print(filtered_data)
```

```python
# Step 7: Select Columns
selected_columns = df[['ObservationDate', 'Confirmed']]
print("\nSelected Columns (Date & Confirmed):\n")
print(selected_columns)
```

```python
# Step 8: Slice First 20 Rows
subset_data = df.iloc[0:20]
print("\nFirst 20 Rows:\n")
print(subset_data)

```

```python
# Step 9: Save Filtered Data
filtered_data.to_csv("filtered_covid_data.csv", index=False)
filtered_data.to_excel("filtered_covid_data.xlsx", index=False)
print("\nFiltered files saved successfully!")

```

```python
# Step 10: Basic Analysis
total_confirmed = df['Confirmed'].sum()
total_deaths = df['Deaths'].sum()
total_recovered = df['Recovered'].sum()

print("\nOverall Covid Summary:")
print("Total Confirmed:", total_confirmed)
print("Total Deaths:", total_deaths)
print("Total Recovered:", total_recovered)

```

```python
# Step 11: Find Highest Cases Day
max_confirmed = df.loc[df['Confirmed'].idxmax()]

print("\nDay with Highest Confirmed Cases:\n")
print(max_confirmed)
```
